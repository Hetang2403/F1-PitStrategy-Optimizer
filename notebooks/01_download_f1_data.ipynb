```python
import fastf1
from pathlib import Path

# Enable cache to speed up data loading
Path("data/cache").mkdir(parents=True, exist_ok=True)
fastf1.Cache.enable_cache('data/cache')

print("✅ FastF1 imported and cache enabled!")
```

```python
print("downloading 2023 Baharain Grand Prix data...")
session = fastf1.get_session(2023, 'Bahrain', 'R')
session.load()

print("✅ Data downloaded!")
print(f"Races date:", session.date)
print(f"Race:", session.event['EventName'])
```

```python
laps = session.laps

pit_stops = laps[laps['PitOutTime'].notna()]

print(f"Total laps: {len(laps)}")
print (f"Total pit stops: {len(pit_stops)}")
print(pit_stops[['Driver', 'LapNumber', 'PitInTime', 'PitOutTime', 'Compound']].head())
```

```python
all_pit_stops = []
for race_round in range(1,23):
    try:
        session = fastf1.get_session(2023, race_round, "R")
        session.load()

        laps = session.laps
        race_pit_stops = laps[laps['PitOutTime'].notna()]
        race_pit_stops['race_round'] = race_round
        race_pit_stops['RaceName'] = session.event['EventName']
        all_pit_stops.append(race_pit_stops)
        print(f"{session.event['EventName']}: {len(race_pit_stops)} pit stops")
    except Exception as e:
        print("failed")

print("download complete")
```

```python
# Combine all races into one big dataset
import pandas as pd
print("🔗 Combining all races into one dataset...")

combined_data = pd.concat(all_pit_stops, ignore_index=True)

print(f"✅ Combined successfully!")
print(f"\n📈 Total pit stops from all 2023 races: {len(combined_data)}")
print(f"\n🔍 First few rows:")
print(combined_data[['Driver', 'LapNumber', 'Compound', 'race_round', 'RaceName']].head())
```

```python
# Save our 2023 data
output_file = 'data/raw/f1_2023_pit_stops.csv'
combined_data.to_csv(output_file, index=False)

print(f"✅ Saved to: {output_file}")
print(f"📊 Total pit stops saved: {len(combined_data)}")
```

```python
import fastf1
import pandas as pd
import time

print("🏎️  MULTI-SEASON F1 DATA DOWNLOADER")
print("=" * 70)

# Store all seasons
all_seasons_data = []

# Define seasons and their race counts
seasons = {
    2020: 17,  # COVID-shortened season
    2021: 22,  # Full season
    2022: 22,  # Full season
    2024: 24   # Try all, will fail for future races (that's ok)
}

# Loop through each season
for year, expected_races in seasons.items():
    print(f"\n{'='*70}")
    print(f"📅 SEASON {year}")
    print(f"{'='*70}")
    
    season_pit_stops = []
    successful_races = 0
    
    # Loop through all races in the season
    for race_round in range(1, expected_races + 1):
        try:
            print(f"\n   Race {race_round:2d}/{expected_races}...", end=" ")
            
            # Load the race session
            session = fastf1.get_session(year, race_round, 'R')
            session.load()
            
            # Get laps data
            laps = session.laps
            
            # Filter for pit stops only
            race_pit_stops = laps[laps['PitOutTime'].notna()].copy()
            
            # Add metadata
            race_pit_stops['Year'] = year
            race_pit_stops['RaceRound'] = race_round
            race_pit_stops['RaceName'] = session.event['EventName']
            
            # Add to collection
            season_pit_stops.append(race_pit_stops)
            successful_races += 1
            
            print(f"✅ {session.event['EventName']:30s} → {len(race_pit_stops):3d} pit stops")
            
            # Small delay to be nice to servers
            time.sleep(0.5)
            
        except Exception as e:
            print(f"❌ Skipped (race not available yet or error)")
            # For 2024, races might not have happened yet - that's ok!
            continue
    
    # Combine all races from this season
    if season_pit_stops:
        season_df = pd.concat(season_pit_stops, ignore_index=True)
        all_seasons_data.append(season_df)
        
        print(f"\n   🎉 {year} COMPLETE!")
        print(f"      Races processed: {successful_races}/{expected_races}")
        print(f"      Total pit stops: {len(season_df):,}")
    else:
        print(f"\n   ⚠️  No data collected for {year}")

print(f"\n{'='*70}")
print("✅ ALL SEASONS DOWNLOADED!")
print(f"{'='*70}")

# Combine all seasons
if all_seasons_data:
    all_years_combined = pd.concat(all_seasons_data, ignore_index=True)
    
    print(f"\n📊 COMPLETE DATASET SUMMARY:")
    print(f"   Total pit stops: {len(all_years_combined):,}")
    print(f"\n   Breakdown by year:")
    for year in sorted(all_years_combined['Year'].unique()):
        count = (all_years_combined['Year'] == year).sum()
        races = all_years_combined[all_years_combined['Year'] == year]['RaceRound'].nunique()
        print(f"      {year}: {count:4,} pit stops across {races:2d} races")
    
    # Save combined dataset
    output_file = 'data/raw/f1_multi_season_pit_stops.csv'
    all_years_combined.to_csv(output_file, index=False)
    print(f"\n💾 Saved to: {output_file}")
    print(f"   File size: {len(all_years_combined):,} rows × {len(all_years_combined.columns)} columns")
    
else:
    print("\n❌ No data was collected!")

print(f"\n🎯 Ready for feature engineering!")
```

```python
print("\n🔗 COMBINING 2023 DATA WITH MULTI-SEASON DATA")
print("=" * 70)

# Load 2023 data (original)
df_2023 = pd.read_csv('data/raw/f1_2023_pit_stops.csv')
print(f"📊 2023 data: {len(df_2023)} pit stops")

# Load multi-season data (2020, 2021, 2024)
df_multi = pd.read_csv('data/raw/f1_multi_season_pit_stops.csv')
print(f"📊 Multi-season data: {len(df_multi)} pit stops")

# Add Year column to 2023 data if it doesn't have it
if 'Year' not in df_2023.columns:
    df_2023['Year'] = 2023
    print("   ✅ Added Year=2023 to 2023 data")

# Find common columns (both datasets must have same columns)
common_columns = list(set(df_2023.columns) & set(df_multi.columns))
print(f"\n📋 Common columns: {len(common_columns)}")

# Keep only common columns in both datasets
df_2023_aligned = df_2023[common_columns].copy()
df_multi_aligned = df_multi[common_columns].copy()

# Combine all data
df_complete = pd.concat([df_2023_aligned, df_multi_aligned], ignore_index=True)

print(f"\n✅ COMBINED DATASET:")
print(f"   Total pit stops: {len(df_complete):,}")

# Show breakdown by year
print(f"\n📅 Breakdown by year:")
for year in sorted(df_complete['Year'].unique()):
    count = (df_complete['Year'] == year).sum()
    races = df_complete[df_complete['Year'] == year]['RaceRound'].nunique() if 'RaceRound' in df_complete.columns else 'N/A'
    print(f"      {year}: {count:4,} pit stops across {races:2} races")

# Save complete dataset
output_file = 'data/raw/f1_complete_2020_2024.csv'
df_complete.to_csv(output_file, index=False)

print(f"\n💾 Saved complete dataset!")
print(f"   File: {output_file}")
print(f"   Size: {len(df_complete):,} rows × {len(df_complete.columns)} columns")

print(f"\n🎯 This is your COMPLETE dataset for feature engineering!")
```
